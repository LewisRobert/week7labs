In this assignment I am currently working in the Information Systems department for a consulting firm working with the state government agency that oversees the healthcare system in California. Here, I will merge the data, export the data, rename columns, insert records, filter data, and convert date-time data.

Importing Datasets and Libraries 

```python
import pandas as pd
from datetime import datetime
```

```python
# Load datasets
hospital_df = pd.read_csv("data/CaliforniaHospitalData.csv")
personnel_df = pd.read_csv("data/CaliforniaHospitalData_Personnel.txt", delimiter="\t")
```

```python
hospital_df.head() #Inspecting Data
```

```python
personnel_df.head() #Inspecting Data
```

Merging the Data

```python
merged_df = pd.merge(personnel_df, hospital_df, on='HospitalID', how='left')
```

```python
#Dropping Unneeded Columns 
merged_df.drop(columns=['Work_ID', 'PositionID', 'Website'], inplace=True)
```

```python
merged_df.head() #Checking the data
```

Filter and Exporting the Data

```python
filtered_df = merged_df[
    (merged_df['Teaching'] == 'Small/Rural') &
    (merged_df['AvlBeds'] >= 15) &
    (merged_df['OperInc'] >= 0)
]

# Export as tab-delimited (no index)
filtered_df.to_csv("data/hospital_data_new.txt", sep='\t', index=False)
```

```python
filtered_df.head() #Checking the data
```

Reloading and Renaming Columns

```python
#reloading into new dataframe
new_df = pd.read_csv("data/hospital_data_new.txt", delimiter="\t")

# Renaming columns
new_df.rename(columns={
    'NoFTE': 'FullTimeCount',
    'NetPatRev': 'NetPatientRevenue',
    'InOperExp': 'InpatientOperExp',
    'OutOperExp': 'OutpatientOperExp',
    'OperRev': 'Operating_Revenue',
    'OperInc': 'Operating_Income'
}, inplace=True)
```

Inserting Records

```python
# Picking two hospitals by HospitalID
hosp1 = new_df.iloc[0].copy()
hosp2 = new_df.iloc[1].copy()

# Inserting myself into the data
new_rows = pd.DataFrame([
    {
        **hosp1,
        'FirstName': 'Lewis',
        'LastName': 'Hughes',
        'PositionTitle': 'Regional Representative',
        'Compensation': 46978,
        'MaxTerm': 4,
        'StartDate': '4/14/25'
    },
    {
        **hosp2,
        'FirstName': 'Lewis',
        'LastName': 'YourLastName',
        'PositionTitle': 'Acting Director',
        'Compensation': 248904,
        'MaxTerm': 8,
        'StartDate': '4/14/25'
    }
])

# Appending using concat
new_merge = pd.concat([new_df, new_rows], ignore_index=True)
```

```python
new_merge.head() #Checking the Data
```

Filtering Data

Non-Profit Filter

```python
nonprofit_filtered = new_merge[
    (new_merge['TypeControl'] == 'Non Profit') &
    (new_merge['FullTimeCount'] > 250) &
    (new_merge['NetPatientRevenue'] >= 109000)
].drop(columns=['FirstName', 'LastName', 'PositionTitle', 'Compensation', 'MaxTerm', 'StartDate'])
```

```python
nonprofit_filtered.head()
```

Regional Representative filter

```python
rep_filtered = new_merge[
    (new_merge['PositionTitle'] == 'Regional Representative') &
    (new_merge['Operating_Income'] > 100000)
]
```

```python
rep_filtered.head()
```

Convert Date-Time Data

```python
new_merge['StartDate'] = pd.to_datetime(new_merge['StartDate'], errors='coerce')

# Confirm data types
print(new_merge.dtypes)

# Preview converted column
print(new_merge['StartDate'].head())
```
